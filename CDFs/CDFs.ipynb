```python
import numpy as np
import matplotlib.pyplot as plt
import cdfs
```

```python
arr = np.random.normal(1,0.2,1000)
```

```python
plt.hist(arr,bins=30,color='k')
```

```python
x, y = cdfs.CDF(arr,norm=False)
```

```python
plt.plot(x,y)
```

```python
x2, y2 = cdfs.CDF(arr,rng=[0,2])
```

```python
plt.plot(x2,y2)
```
